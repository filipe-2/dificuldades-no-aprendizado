```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder

# Array de perguntas
questions = [
    "Nome",
    "Idade",
    "Sexo",
    "Trabalha",
    "Área de estudos",
    "Maior dificuldade nos estudos",
    "Possui distúrbio diagnosticado",
    "Emprego atrapalha estudos",
    "Redes sociais atrapalham estudos",
    "Problemas pessoais atrapalham estudos",
    "Ainda possui interesse na área de estudos atual"
]

# Array de respostas
possible_answers = [
    None,  # Nome completo - resposta por input
    None,  # Idade - resposta por input
    ['Masculino', 'Feminino'],  # Sexo
    ['Sim', 'Nao'],  # Você trabalha?
    ['Tecnologia', 'Humanas', 'Ciencias Naturais', 'Ciencias Exatas', 'Outra'],  # Área de estudo
    ['Memorizacao', 'Tempo', 'Capacidade cognitiva reduzida', 'Foco/atencao', 'Interesse', 'Outra'],  # Dificuldades?
    ['Sim', 'Nao'],  # Transtornos atrapalham estudos?
    ['Sim', 'Nao'],  # Trabalho atrapalha estudos?
    ['Sim', 'Nao'],  # Redes sociais atrapalham estudos?
    ['Sim', 'Nao'],  # Problemas pessoais atrapalham estudos?
    ['Sim', 'Nao']  # Interesse na área de estudos atual?
]

# Gerar respostas aleatórias
num_rows = 100
data = {}

for i, (question, possible_answers) in enumerate(zip(questions, possible_answers)):
    if possible_answers is None:
        data[question] = np.random.choice(['Aleatorio ' + str(i) for i in range(100)], size=num_rows)
    elif question == "Qual sua idade?":
        data[question] = np.random.randint(13, 85, size=num_rows)  # Gerar idades entre 13 e 85 anos
    else:
        data[question] = np.random.choice(possible_answers, size=num_rows)

# Criar DataFrame
df = pd.DataFrame(data)

# Converter respostas binárias para 0 e 1
df.replace({'Sim': 1, 'Nao': 0}, inplace=True)

# Escrever dados para CSV
df.to_csv('random_answers.csv', index=False)
```
