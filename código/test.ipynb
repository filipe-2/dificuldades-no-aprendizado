```python
import pandas as pd
import numpy as np

# Array de perguntas
questions = [
    "Qual seu nome completo?",
    "Qual sua idade?",
    "Voce trabalha?",
    "Qual sua area de estudos?",
    "Em qual dos seguintes itens voce possui mais dificuldade durante seu aprendizado?",
    "Voce possui algum disturbio ou transtorno que afeta suas capacidades cognitivas?",
    "Voce considera que seu emprego atual pode estar atrapalhando seus estudos?",
    "Voce considera que suas redes sociais podem estar atrapalhando seus estudos?",
    "Voce acredita que seus problemas pessoais?",
    "Voce ainda possui interesse na sua area de estudo atual?"
]

# Array de respostas
possible_answers = [
    None,  # Nome completo - resposta por input
    None,  # Idade - resposta por input
    ['Sim', 'Nao'],  # Você trabalha?
    ['Tecnologia', 'Humanas', 'Ciencias Naturais', 'Ciencias Exatas', 'Outra'],  # Área de estudo
    ['Memorizacao', 'Tempo', 'Capacidade cognitiva reduzida', 'Foco/atencao', 'Interesse', 'Outra'],  # Dificuldades?
    ['Sim', 'Nao', "Nao sei"],  # Transtornos atrapalham estudos?
    ['Sim', 'Nao'],  # Trabalho atrapalha estudos?
    ['Sim', 'Nao'],  # Redes sociais atrapalham estudos?
    ['Sim', 'Nao'],  # Problemas pessoais atrapalham estudos?
    ['Sim', 'Nao']  # Interesse na área de estudos atual?
]

# Gerar respostas aleatórias
num_rows = 100
data = {}

for i, (question, possible_answers) in enumerate(zip(questions, possible_answers)):
    if question == "Qual sua idade?":
        data[question] = np.random.randint(18, 70, size=num_rows)  # Generate random ages between 18 and 70
    elif possible_answers is None:
        data[question] = np.random.choice(['Aleatorio ' + str(i) for i in range(100)], size=num_rows)
    else:
        data[question] = np.random.choice(possible_answers, size=num_rows)

# Criar DataFrame
df = pd.DataFrame(data)

# Converter respostas binárias para 0 e 1
df.replace({'Sim': 1, 'Nao': 0, "Eu nao sei": np.nan}, inplace=True)

# Escrever dados para CSV
df.to_csv('random_answers.csv', index=False)
```
